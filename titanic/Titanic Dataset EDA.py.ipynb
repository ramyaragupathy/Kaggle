## Plot a graph to understand who has survived

1 = Survived | 0 = Not Survived

- First graph shows the percentage: 40% survived
- Second graph shows the number: 350 people survived

```python
fig  = plt.figure(figsize=(18,6))
plt.subplot2grid((3,3),(0,0))
df.Survived.value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plt.title("40% of people survived")
plt.subplot2grid((3,3),(0,1))
df.Survived.value_counts().plot(kind="bar", alpha=0.5)
plt.title("350 people survived")
plt.subplot2grid((3,3),(0,2))
plt.scatter(df.Survived, df.Age, alpha = 0.1)
plt.title("Age wrt Survived")
plt.subplot2grid((3,3),(1,0))
df.Pclass.value_counts().plot(kind="bar", alpha=0.5)
plt.title("Class")
plt.subplot2grid((3,3),(1,1))
df.Pclass.value_counts(normalize=True).plot(kind="bar", alpha=0.5)
plt.title("Class")
plt.subplot2grid((3,3),(1,2))
plt.scatter(df.Survived, df.Pclass, alpha = 0.1)
plt.title("Class wrt Survived")
plt.subplot2grid((3,3),(2,0), colspan=2)
for x in [1,2,3]:
    df.Age[df.Pclass == x].plot(kind="kde")
plt.title("Class wrt Age")
plt.legend(("1st class", "2nd class", "3rd class"))
plt.subplot2grid((3,3),(2,2))
df.Embarked.value_counts().plot(kind="bar", alpha=0.5)
plt.title("Embarked")
plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=0.5, wspace=0.4)
plt.show()
```
